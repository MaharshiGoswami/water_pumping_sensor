### ----Checking Outliers before filling moderate and low missing values in rest of the columns----

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
sns.boxplot(data=wp[['sensor_00','sensor_06','sensor_07','sensor_08','sensor_09']])
plt.xticks(rotation=45)
plt.show()
```

### We got extreme Class imbalance . since, normal status is 93% and broken status almost zero, model will make everything normal and ignore broken status. so, the model will become useless

## ----Combining Classes----

```python
wp['machine_status']=wp['machine_status'].replace({
    'RECOVERING': 'BROKEN'
})
```

### Now Normal states means "working" and Broken states means "problem"

```python
wp['machine_status'].unique()
```

## ----Encoding----

```python
wp['machine_status']=wp['machine_status'].map({
    'NORMAL':0,
    'BROKEN':1
})
```

```python
wp.head()
```

## ----Removing TimeStamp----

```python
wp=wp.drop(columns=['timestamp'])
```

## ----Feature Selection----

```python
x=wp.drop(columns=['machine_status'])
y=wp['machine_status']
```

```python
x.columns
```

## ----Train Test Split----

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Since the dataset contains outliers that represent real anomalies, I used RobustScaler to scale the data without being affected by extreme values

# Model Training

## Logistic Regressi

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, Y_train)

y_pred_lr = lr.predict(X_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, Y_train)

y_pred_rf = rf.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Logistic Regression")
print(confusion_matrix(Y_test, y_pred_lr))
print(classification_report(Y_test, y_pred_lr))

print("Random Forest")
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(class_weight='balanced', random_state=42)
rf.fit(X_train, Y_train)
y_pred_rf = rf.predict(X_test)
print(confusion_matrix(Y_test, y_pred_rf))
print(classification_report(Y_test, y_pred_rf))
```

### Here FN:60 in Logistic , it means it failed to detect 60 broken


# Logistic Regression VS Random Forest Confusion Matrix Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Confusion matrices
cm_lr = confusion_matrix(Y_test, y_pred_lr)
cm_rf = confusion_matrix(Y_test, y_pred_rf)

# Subplots
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Logistic Regression
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("Logistic Regression")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

# Random Forest
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title("Random Forest")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```
